1. VendorID : Mã số của nhà cung cấp công nghệ đã ghi nhận chuyến đi. 1 = Creative Mobile Technologies, 2 = Curb Mobility, 6 = Myle Technologies Inc, 7 = Helix.

2.tpep_pickup_datetime: Ngày và giờ đồng hồ tính tiền được kích hoạt (bắt đầu chuyến đi).

3. tpep_dropoff_datetime: Ngày và giờ đồng hồ tính tiền ngừng hoạt động (kết thúc chuyến đi).

4. passenger_count: Số lượng hành khách trên xe (do tài xế khai báo). Lưu ý: có thể chứa giá trị 0 (lỗi nhập liệu) hoặc các giá trị ngoại lai như 9.

5. trip_distance: Quãng đường đã đi được tính bằng dặm, do đồng hồ tính tiền ghi nhận.

6. RatecodeID: Mã loại giá cước cuối cùng của chuyến đi: 1 = Standard, 2 = JFK, 3 = Newark, 4 = Nassau/Westchester, 5 = Thương lượng, 6 = Đi chung, 99 = Không rõ.

7. store_and_fwd_flag: Cờ "lưu trữ và chuyển tiếp": "Y" nếu dữ liệu được lưu trong bộ nhớ xe trước khi gửi về máy chủ (do mất kết nối), "N" nếu gửi trực tiếp.

8. PULocationID: Mã số khu vực taxi (Taxi Zone) nơi khách lên xe.

9. DOLocationID: Mã số khu vực taxi nơi khách xuống xe.

10. payment_type: Mã phương thức thanh toán: 0 = Flex Fare, 1 = Credit Card, 2 = Cash, 3 = No charge, 4 = Dispute, 5 = Unknown, 6 = Voided trip.

11. fare_amount: Số tiền cước phí dựa trên thời gian và quãng đường tính bởi đồng hồ.

12. extra: Các khoản phụ thu và phụ phí linh tinh.

13. mta_tax: Thuế MTA (Cơ quan Giao thông Đô thị) được tự động kích hoạt dựa trên mức cước đang sử dụng.

14. tip_amount: Số tiền tip. Chỉ tự động ghi nhận đối với tip qua thẻ tín dụng. Tip tiền mặt không được bao gồm.

15. tolls_amount: Tổng số tiền phí cầu đường đã trả trong chuyến đi.

16. improvement_surcharge: Phụ phí cải thiện taxi ($0.30 cho mỗi chuyến) bắt đầu từ năm 2015 để hỗ trợ cải thiện khả năng tiếp cận.

17. total_amount: Tổng số tiền khách trả, không bao gồm tip tiền mặt.

18. congestion_surcharge: Phụ phí ùn tắc giao thông của bang New York.

19. airport_fee: Phí sân bay, chỉ áp dụng khi đón khách tại LaGuardia và JFK.

```python
no_passenger = df[df['passenger_count']==0].shape[0]
print(f"So chuyen khong co khach: {no_passenger}")
```

```python
max_row = df.loc[df['total_amount'].idxmax()]
print("Chuyen co tong tien cao nhat:")
print(max_row[['tpep_pickup_datetime','trip_distance','total_amount']])
```

```python
# Lọc các chuyến đón ở Times Square
ts_pickups = df[df['PULocationID'] == 230]
avg_distance_ts = ts_pickups['trip_distance'].mean()
print(f"Quãng đường trung bình từ Times Square: {avg_distance_ts:.2f} dặm")
```

```python
df['trip_duration_minutes'] = (df['tpep_dropoff_datetime']-df['tpep_pickup_datetime']).dt.total_seconds()/60

valid_trips = df[(df['trip_duration_minutes']>0)&(df['trip_duration_minutes']<180)]

print(f"Trung binh(mean): {valid_trips['trip_duration_minutes'].mean():.1f} phut")
print(f"Trung vi(median): {valid_trips['trip_duration_minutes'].median():.1f} phut")
```

```python
# Tạo cột giờ và ngày trong tuần
df['pickup_hour'] = df['tpep_pickup_datetime'].dt.hour
df['pickup_dayofweek'] = df['tpep_pickup_datetime'].dt.dayofweek  # 0=Thứ Hai, 6=Chủ Nhật

# Số chuyến theo giờ
hourly_counts = df.groupby('pickup_hour').size().sort_values(ascending=False)
print("Top 5 khung giờ đông nhất:")
print(hourly_counts.head(5))

# Bonus: Số chuyến theo giờ và ngày trong tuần (pivot table)
day_hour_pivot = df.pivot_table(index='pickup_dayofweek', columns='pickup_hour', aggfunc='size')
print("\nHeatmap sơ bộ (hàng=thứ, cột=giờ):")
print(day_hour_pivot)
```

```python
# Tạo cột tip_percent
df['tip_percent'] = (df['tip_amount'] / df['fare_amount']) * 100

# Chỉ lấy thanh toán bằng thẻ để so sánh tip
card_trips = df[(df['payment_type'] == 1) & (df['fare_amount'] > 0)]

# Phân loại giờ cao điểm sáng
card_trips['is_rush_hour'] = card_trips['pickup_hour'].isin([7, 8, 9])

rush_tip = card_trips[card_trips['is_rush_hour']]['tip_percent'].median()
offpeak_tip = card_trips[~card_trips['is_rush_hour']]['tip_percent'].median()

print(f"Tip trung vị giờ cao điểm sáng (7-9AM): {rush_tip:.1f}%")
print(f"Tip trung vị giờ thấp điểm: {offpeak_tip:.1f}%")
```

Có 5 cột có dữ liệu null: 71743 giá trị
1. passenger_count: Số lượng hành khách trên xe
2. RatecodeID: Mã loại giá cước cuối cùng của chuyến đi
3. store_and_fwd_flag: Cờ "lưu trữ và chuyển tiếp"
4. congestion_surcharge: Phụ phí ùn tắc giao thông của bang New York.
5. airport_fee: Phí sân bay, chỉ áp dụng khi đón khách tại LaGuardia và JFK.

```python
# Xóa toàn bộ các dòng có giá trị null ở cột 'passenger_count'
df_clean = df.dropna(subset=['passenger_count']).copy()
```

```python
# Tìm giá trị xuất hiện nhiều nhất của cột 'RatecodeID'
mode_ratecode = df_clean['RatecodeID'].mode()[0]
# Thay thế các giá trị null bằng giá trị mode vừa tìm được
df_clean['RatecodeID'] = df_clean['RatecodeID'].fillna(mode_ratecode)
```

```python
# Xóa cột 'store_and_fwd_flag' khỏi DataFrame
df_clean = df_clean.drop(columns=['store_and_fwd_flag'])
```

```python
# Thay thế giá trị null bằng 0
df_clean['congestion_surcharge'] = df_clean['congestion_surcharge'].fillna(0)
df_clean['airport_fee'] = df_clean['airport_fee'].fillna(0)
```
