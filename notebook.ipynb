1. VendorID : Mã số của nhà cung cấp công nghệ đã ghi nhận chuyến đi. 1 = Creative Mobile Technologies, 2 = Curb Mobility, 6 = Myle Technologies Inc, 7 = Helix.

2.tpep_pickup_datetime: Ngày và giờ đồng hồ tính tiền được kích hoạt (bắt đầu chuyến đi).

3. tpep_dropoff_datetime: Ngày và giờ đồng hồ tính tiền ngừng hoạt động (kết thúc chuyến đi).

4. passenger_count: Số lượng hành khách trên xe (do tài xế khai báo). Lưu ý: có thể chứa giá trị 0 (lỗi nhập liệu) hoặc các giá trị ngoại lai như 9.

5. trip_distance: Quãng đường đã đi được tính bằng dặm, do đồng hồ tính tiền ghi nhận.

6. RatecodeID: Mã loại giá cước cuối cùng của chuyến đi: 1 = Standard, 2 = JFK, 3 = Newark, 4 = Nassau/Westchester, 5 = Thương lượng, 6 = Đi chung, 99 = Không rõ.

7. store_and_fwd_flag: Cờ "lưu trữ và chuyển tiếp": "Y" nếu dữ liệu được lưu trong bộ nhớ xe trước khi gửi về máy chủ (do mất kết nối), "N" nếu gửi trực tiếp.

8. PULocationID: Mã số khu vực taxi (Taxi Zone) nơi khách lên xe.

9. DOLocationID: Mã số khu vực taxi nơi khách xuống xe.

10. payment_type: Mã phương thức thanh toán: 0 = Flex Fare, 1 = Credit Card, 2 = Cash, 3 = No charge, 4 = Dispute, 5 = Unknown, 6 = Voided trip.

11. fare_amount: Số tiền cước phí dựa trên thời gian và quãng đường tính bởi đồng hồ.

12. extra: Các khoản phụ thu và phụ phí linh tinh.

13. mta_tax: Thuế MTA (Cơ quan Giao thông Đô thị) được tự động kích hoạt dựa trên mức cước đang sử dụng.

14. tip_amount: Số tiền tip. Chỉ tự động ghi nhận đối với tip qua thẻ tín dụng. Tip tiền mặt không được bao gồm.

15. tolls_amount: Tổng số tiền phí cầu đường đã trả trong chuyến đi.

16. improvement_surcharge: Phụ phí cải thiện taxi ($0.30 cho mỗi chuyến) bắt đầu từ năm 2015 để hỗ trợ cải thiện khả năng tiếp cận.

17. total_amount: Tổng số tiền khách trả, không bao gồm tip tiền mặt.

18. congestion_surcharge: Phụ phí ùn tắc giao thông của bang New York.

19. airport_fee: Phí sân bay, chỉ áp dụng khi đón khách tại LaGuardia và JFK.

```python
df['trip_duration_minutes'] = (df['tpep_dropoff_datetime']-df['tpep_pickup_datetime']).dt.total_seconds()/60

valid_trips = df[(df['trip_duration_minutes']>0)&(df['trip_duration_minutes']<180)]

print(f"Trung binh(mean): {valid_trips['trip_duration_minutes'].mean():.1f} phut")
print(f"Trung vi(median): {valid_trips['trip_duration_minutes'].median():.1f} phut")
```

```python
# Tạo cột giờ và ngày trong tuần
df['pickup_hour'] = df['tpep_pickup_datetime'].dt.hour
df['pickup_dayofweek'] = df['tpep_pickup_datetime'].dt.dayofweek  # 0=Thứ Hai, 6=Chủ Nhật

# Số chuyến theo giờ
hourly_counts = df.groupby('pickup_hour').size().sort_values(ascending=False)
print("Top 5 khung giờ đông nhất:")
print(hourly_counts.head(5))

# Bonus: Số chuyến theo giờ và ngày trong tuần (pivot table)
day_hour_pivot = df.pivot_table(index='pickup_dayofweek', columns='pickup_hour', aggfunc='size')
print("\nHeatmap sơ bộ (hàng=thứ, cột=giờ):")
print(day_hour_pivot)
```

```python
# Tạo cột tip_percent
df['tip_percent'] = (df['tip_amount'] / df['fare_amount']) * 100

# Chỉ lấy thanh toán bằng thẻ để so sánh tip
card_trips = df[(df['payment_type'] == 1) & (df['fare_amount'] > 0)]

# Phân loại giờ cao điểm sáng
card_trips['is_rush_hour'] = card_trips['pickup_hour'].isin([7, 8, 9])

rush_tip = card_trips[card_trips['is_rush_hour']]['tip_percent'].median()
offpeak_tip = card_trips[~card_trips['is_rush_hour']]['tip_percent'].median()

print(f"Tip trung vị giờ cao điểm sáng (7-9AM): {rush_tip:.1f}%")
print(f"Tip trung vị giờ thấp điểm: {offpeak_tip:.1f}%")
```

Quyết định khi xử lý giá trị missing:

1. Tầm quan trọng của dữ liệu đối với mục đích dự đoán

2. Việc thay thế bằng các giá trị khác có hợp lí hay không? Có làm ảnh hưởng tới chính xác và kết quả sau này của bài toán?

3. Rủi ro khi mất mát dữ liệu đó là gì?

=> Từ những điều trên ta mới có thể quyết định nên xóa hay thay thế các dữ liệu missing.

=> Cần ghi nhớ: luôn phải đặt câu hỏi là “giá trị của cột dữ liệu này đối với bài toán là gì?”

```python
df_clean['tpep_pickup_datetime'] = pd.to_datetime(df_clean['tpep_pickup_datetime'])
df_clean['tpep_dropoff_datetime'] = pd.to_datetime(df_clean['tpep_dropoff_datetime'])

df_clean['trip_duration_minutes'] = (df_clean['tpep_dropoff_datetime'] - df_clean['tpep_pickup_datetime']).dt.total_seconds() / 60
```

```python
df_clean[['trip_duration_minutes', 'trip_distance', 'passenger_count', 'fare_amount', 'total_amount']].describe()
```

```python
mask = pd.Series(True ,index = df_clean.index)

# 0 < thoi gian chuyen di <= 180
mask = mask & (df_clean['trip_duration_minutes']>0) & (df_clean['trip_duration_minutes']<=180)

# khoang cach > 0
mask = mask & (df_clean['trip_distance'] > 0)

# 1 <= so hanh khach <= 7 
mask = mask & (df_clean['passenger_count'] >=1) & (df_clean['passenger_count']<=7)

# gia ve co ban > 0
mask = mask & (df_clean['fare_amount'] > 0)

# tong tien > 0
mask = mask & (df_clean['total_amount'] > 0)
```

```python
df_clean_no_outliers = df_clean[mask].copy()
```

```python
print(f"Số dòng trước lọc: {len(df_clean)}")
print(f"Số dòng sau khi lọc: {len(df_clean_no_outliers)}")
print(f"Số dòng bị loại: {len(df_clean)-len(df_clean_no_outliers)}")
print(f"Tỷ lệ loại: {(len(df_clean) - len(df_clean_no_outliers)) / len(df_clean) *100:.2f}%")
```

```python
print("Số dòng vi phạm từng điều kiện (có thể trùng lặp):")
print(f"  duration <= 0 hoặc > 180: {len(df_clean[~((df_clean['trip_duration_minutes'] > 0) & (df_clean['trip_duration_minutes'] <= 180))])}")
print(f"  trip_distance <= 0: {len(df_clean[df_clean['trip_distance'] <= 0])}")
print(f"  passenger_count < 1 hoặc > 7: {len(df_clean[(df_clean['passenger_count'] < 1) | (df_clean['passenger_count'] > 7)])}")
print(f"  fare_amount <= 0: {len(df_clean[df_clean['fare_amount'] <= 0])}")
print(f"  total_amount <= 0: {len(df_clean[df_clean['total_amount'] <= 0])}")
```

```python
df_clean_no_outliers['hour'] = df_clean_no_outliers['tpep_pickup_datetime'].dt.hour

df_clean_no_outliers['day_of_week'] = df_clean_no_outliers['tpep_pickup_datetime'].dt.dayofweek

# Tạo cột is_weekend: 1 nếu là Saturday (5) hoặc Sunday (6), else 0
df_clean_no_outliers['is_weekend'] = (df_clean_no_outliers['day_of_week']>=5).astype(int)
```

```python
int_cols = ['passenger_count', 'RatecodeID', 'payment_type', 'hour', 'day_of_week', 'is_weekend', 'PULocationID', 'DOLocationID']

for col in int_cols:
    if col in df_clean_no_outliers.columns:
        df_clean_no_outliers[col] = df_clean_no_outliers[col].astype('int32')

float_cols = ['trip_distance', 'fare_amount', 'extra', 'mta_tax', 'tip_amount', 'tolls_amount', 
              'improvement_surcharge', 'total_amount', 'congestion_surcharge', 'airport_fee', 
              'trip_duration_minutes', 'avg_speed_mph']
for col in float_cols:
    if col in df_clean_no_outliers.columns:
        df_clean_no_outliers[col] = df_clean_no_outliers[col].astype('float32')
```

```python
print(df_clean_no_outliers.info(memory_usage='deep'))
```

```python
# 1. Lọc outlier cho avg_speed_mph (thêm ngay sau cell tính avg_speed_mph)
df_clean_no_outliers = df_clean_no_outliers[df_clean_no_outliers['avg_speed_mph'] <= 100].copy()
print(f"Số dòng sau khi lọc speed > 100 mph: {len(df_clean_no_outliers)}")

# 2. Xóa các cột trùng lặp hoặc không cần thiết
df_clean_no_outliers = df_clean_no_outliers.drop(columns=['pickup_hour', 'pickup_dayofweek', 'tip_percent', 'is_weeked'], errors='ignore')

# 3. Ép kiểu nốt VendorID xuống int8
df_clean_no_outliers['VendorID'] = df_clean_no_outliers['VendorID'].astype('int8')

# 4. Kiểm tra lại memory
print(df_clean_no_outliers.info(memory_usage='deep'))
```

```python
start_date = '2023-01-01'
end_date = '2023-01-31 23:59:59'

df_jan = df_clean_no_outliers[
    (df_clean_no_outliers['tpep_pickup_datetime'] >= start_date) & 
    (df_clean_no_outliers['tpep_pickup_datetime'] <= end_date)
].copy()

print(f"Số dòng trong tháng 1/2023: {len(df_jan)}")
print(f"Min datetime sau lọc: {df_jan['tpep_pickup_datetime'].min()}")
print(f"Max datetime sau lọc: {df_jan['tpep_pickup_datetime'].max()}")
```

```python
df_temp = df_jan.set_index('tpep_pickup_datetime')
```

```python
hourly_data = df_temp.resample('H').agg(
    total_pickups=('trip_distance', 'count'),
    avg_duration_minutes=('trip_duration_minutes', 'mean'),
    avg_distance_miles=('trip_distance', 'mean'),
    avg_fare_amount=('fare_amount', 'mean'),
    avg_total_amount=('total_amount', 'mean'),
    avg_speed_mph=('avg_speed_mph', 'mean')
).reset_index()
```

```python
hourly_data['hour_of_day'] = hourly_data['tpep_pickup_datetime'].dt.hour
hourly_data['day_of_week'] = hourly_data['tpep_pickup_datetime'].dt.dayofweek
hourly_data['is_weekend'] = (hourly_data['day_of_week'] >= 5).astype(int)
```

```python
avg_cols = ['avg_duration_minutes', 'avg_distance_miles', 'avg_fare_amount', 'avg_total_amount', 'avg_speed_mph']
hourly_data[avg_cols] = hourly_data[avg_cols].fillna(0)
```

```python
lags = [1, 2, 3, 24]
for lag in lags:
    hourly_data[f'lag_{lag}'] = hourly_data['total_pickups'].shift(lag)
```

```python
windows = [3, 6, 12, 24]
for w in windows:
    hourly_data[f'roll_mean_{w}'] = hourly_data['total_pickups'].rolling(window=w, min_periods=1).mean().shift(1)
```
